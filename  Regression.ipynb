```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('data/cluster_csv/all_day20201210.csv')
df1
```

```python
#当日朝5時から7時の3時間を入力、その日の最高値or値を出した時刻を出力(svr_model1_seed)
df = df1[1:]
X,Y1,Y2 = np.array([]), np.array([]), np.array([])
h_label = np.array(list(range(24)))
for i in range(len(df)):
    X = np.append(X, df.iloc[i, 6:9].values)
    l = df.iloc[i, 1:].values
    Y1 = np.append(Y1, np.amax(l))
    Y2 = np.append(Y2, h_label[l==np.amax(l)])
X = X.reshape(-1,3)
print(X)
print(Y1)
print(Y2)

#result
# 7677.614974707609
# 8699.257418150093
```

```python
#前日1日のデータを入力、翌日の最高値を出力(svr_model2)
df = df1[1:]
X,Y1,Y2 = np.array([]), np.array([]), np.array([])
h_label = np.array(list(range(24)))
for i in range(len(df)-1):
    l = df.iloc[i, 1:].values
    l2 = df.iloc[i+1, 1:].values
    X = np.append(X, l)
    Y1 = np.append(Y1, np.amax(l2))
    Y2 = np.append(Y2, h_label[l2==np.amax(l2)])
X = X.reshape(-1,24)
print(X)
print(Y1)
print(Y2)

#result
# 4961.349842693643
# 5571.642238314589
#seed2>
# 5142.966095296689
# 5026.793480505451
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
seed = 2
x_train, x_test, y_train, y_test =\
    train_test_split(X, Y1, random_state=seed)
```

```python
len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
import random
```

```python
#モデルの作成　svr
from sklearn import svm
from sklearn.model_selection import cross_val_score

clf = svm.SVR(kernel='linear').fit(x_train,y_train)
n = random.randint(0,100)
print(clf.predict([x_train[n]]),y_train[n])
```

```python
# 教師データに対して検証
# for i in range(len(x_train)):
#     abs(x_train-y_train)/x_train[i]
print(abs(clf.predict(x_train)-y_train).mean())
print(abs(clf.predict(x_test)-y_test).mean())
```

```python
#モデルの保存
import pickle
path = 'model/svr_model1_'+str(seed)+'.sav'
pickle.dump(clf, open(path, 'wb'))
```
